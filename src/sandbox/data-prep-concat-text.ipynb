# Prepare Data

```python
import json
import pandas as pd
import numpy as np

pd.set_option('max_colWidth', 200)
```

```python
# with open('/home/rnam/documents/squad/data/train-v1.1.json') as f:
with open('/Users/richardnam/Google Drive/squad/data/train-v1.1.json') as f:
    data = json.load(f)
```

```python
total = []
for article_id, article in enumerate(_data):
    article_title = article['title']
    for i, v in enumerate(article['paragraphs']):
        contextidx = i
        context = v['context']
        for _i, _v in enumerate(v['qas']):
            answers = _v['answers']
            qid = _v['id']
            question = _v['question']
            for _n in answers:
                answer_start = _n['answer_start']
                answer_text = _n['text']
                row = [article_id, article_title, contextidx, context, qid, question, answer_start, answer_text]
                total.append(row)
```

```python
df = pd.DataFrame(total, columns=['article_id', 'article_title', 'context_id','context',
                                  'question_id','question','answer_start','answer_text'])
```

### Save all the articles (as a whole)

```python
article_ids = df.article_id.unique()
```

```python
for artid in article_ids:
    _dftmp = df[df.article_id==artid]
    _dftmp.drop_duplicates(subset=['context_id'], inplace=True)
    _dftmp.sort_values('context_id', inplace=True)
    _dflen = _dftmp.shape[0]
    _conid = _dftmp.context_id.values
    _con = _dftmp.context.values
    _title = _dftmp.article_title.unique()[0]
    output = open('/Users/richardnam/Google Drive/squad/data/articles/%s.txt' % _title, 'w')
    for i in range(_dflen):
        output.write('%d||%s\n\n' % (_conid[i], _con[i]))
    output.close()
```
